# BMOLM — Chat with BMO

Download a pre-trained 9M parameter BMO LLM and chat with it. Just run all cells.

**Model:** [arman-bd/bmolm-9M](https://huggingface.co/arman-bd/bmolm-9M)

```python
# Setup + Download
!pip install -q torch tokenizers huggingface_hub
import os, shutil
if os.path.exists('/content/bmo'): shutil.rmtree('/content/bmo')
os.makedirs('/content/bmo'); os.chdir('/content/bmo')

from huggingface_hub import snapshot_download
snapshot_download(repo_id='arman-bd/bmolm-9M', local_dir='.')
print('Model downloaded.')
```

```python
# Load model
from inference import BMOInference
import torch

engine = BMOInference('pytorch_model.bin', 'tokenizer.json',
                        device='cuda' if torch.cuda.is_available() else 'cpu')

def chat(prompt):
    return engine.chat_completion(
        [{'role': 'user', 'content': prompt}], max_tokens=64
    )['choices'][0]['message'].get('content', '').strip()

# Quick test
for p in ['hi bmo', 'are you hungry', 'tell me a joke', 'what is the internet', 'goodnight bmo']:
    print(f'You> {p}\nBMO> {chat(p)}\n')
```

```python
# Interactive chat — type your messages
while True:
    try:
        p = input('You> ').strip()
    except (KeyboardInterrupt, EOFError):
        break
    if not p or p.lower() in ('quit', 'exit', 'q'):
        print('BMO> bye. i will continue being a game console.'); break
    print(f'BMO> {chat(p)}\n')
```
